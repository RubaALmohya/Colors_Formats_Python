When printing the images it appears to us in BGR format due to the use of CV2 in python

To display the image in RGB format as seen by the eye, there are more than one way:


```python
##First
image = cv2.imread('/content/download.jpg')
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
```

Only here using CV2 and cvtColor() method we pass the image and conversion format we want

```python
#second
import matplotlib.image as mpimg
image = mpimg.imread('/content/download.jpg')
plt.imshow(image)
plt.show()
```

Only here using a subpackage provided by the matplotlib library which is mpimg

we want to convert our original image from the **BGR color space to gray**

```python
#first
image = cv2.imread('/content/download.jpg')
GRAY_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
GRAY2_image = cv2.cvtColor(GRAY_image,cv2.COLOR_BGR2RGB)
plt.imshow(GRAY2_image)
```

```python
image = cv2.imread('/content/download.jpg')
HSV_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
plt.imshow(HSV_image)
```

```python
image = cv2.imread('/content/download.jpg')
HLS_image = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
plt.imshow(HLS_image)
```

```python
image = cv2.imread('/content/download.jpg')
HLS1 = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
HLS2_image = cv2.cvtColor(HLS1,cv2.COLOR_BGR2RGB)
plt.imshow(HLS2_image)
```

Only here using CV2 and cvtColor() method we pass the image and conversion format we want

```python

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
 

fname = r'download.jpg'
image = Image.open(fname).convert("L")

plt.imshow(image, cmap='gray')
plt.show()
```


we used to display a grayscale image in Matplotlib, we use the matplotlib. pyplot. imshow() with parameters cmap set to 'gray' 

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
 
fname = r'download.jpg'

CMYK_image = Image.open(fname).convert('CMYK')
plt.imshow(CMYK_image)
plt.show()

```

This is the easiest way to convert color modes in Python
